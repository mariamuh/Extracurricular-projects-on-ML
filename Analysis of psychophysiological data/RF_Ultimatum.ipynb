# Используем все предикторы для обучения модели

```python
data.columns
```

```python
X = data[['Participant Private ID', 'Machiavellianism', 'Psychopathy',
       'Narcissism', 'Perspective-taking scale', 'Fantasy scale',
       'Empathic concern scale', 'Personal distress scale',
       'Tolerance to uncertainty', 'Intolerance to uncertainty',
       'Interpersonal intolerance to uncertainty', 'genderRU-quantised',
       'DoBRU-year', 'educationRU-quantised']]

y = data[['consent_fair_offer',
       'consent_conditionally_fair_offer', 'consent_unfair_offer',
       'rejection_fair_offer', 'rejection_conditionally_fair_offer',
       'rejection_unfair_offer']]
#y_targets = np.array_split(y.to_numpy(), y.shape[0])
```

```python
X[X['DoBRU-year'].isnull() == True]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

rfr = RandomForestRegressor(max_depth=2, random_state=0)
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test).round()
```

```python
y_pred
```

```python
mean_squared_error(y_test, y_pred)
```

```python
rfr2 = RandomForestRegressor(n_estimators=80, max_depth=2, random_state=0)
rfr2.fit(X_train, y_train)

y_pred2 = rfr2.predict(X_test).round()
mean_squared_error(y_test, y_pred2)
```

```python
rfr3 = RandomForestRegressor(n_estimators=80, max_depth=2, max_leaf_nodes=3, random_state=0)
rfr3.fit(X_train, y_train)

y_pred3 = rfr3.predict(X_test).round()
mean_squared_error(y_test, y_pred3)
```

```python
rfr4 = RandomForestRegressor(n_estimators=80, max_depth=2, max_leaf_nodes=2, random_state=0)
rfr4.fit(X_train, y_train)

y_pred4 = rfr4.predict(X_test).round()
mean_squared_error(y_test, y_pred4)
```

```python
rfr4_predictions = rfr4.predict(X).round()
```

```python
rfr4_columns_names = ['rfr_consent_fair_offer',
       'rfr_consent_conditionally_fair_offer', 'rfr_consent_unfair_offer',
       'rfr_rejection_fair_offer', 'rfr_rejection_conditionally_fair_offer',
       'rfr_rejection_unfair_offer']

data_rfr = pd.concat([data, pd.DataFrame(rfr4_predictions, columns=rfr4_columns_names)], axis=1)
data_rfr.head(5)
```

```python
data_rfr.drop(data_rfr.tail(1).index,inplace=True)
```

```python
y_sqrt = np.sqrt(y)
```

```python
X_train_sqrt, X_test_sqrt, y_train_sqrt, y_test_sqrt = train_test_split(
    X, y_sqrt, test_size=0.3, random_state=42)

rfr_sqrt = RandomForestRegressor(max_depth=2, random_state=0)
rfr_sqrt.fit(X_train_sqrt, y_train_sqrt)

y_pred_sqrt = rfr_sqrt.predict(X_test_sqrt).round()
mean_squared_error(y_test_sqrt, y_pred_sqrt)
```

```python
rfr_sqrt2 = RandomForestRegressor(n_estimators=80, max_depth=2, max_leaf_nodes=2, random_state=0)
rfr_sqrt2.fit(X_train_sqrt, y_train_sqrt)

y_pred_sqrt2 = rfr_sqrt2.predict(X_test_sqrt).round()
mean_squared_error(y_test_sqrt, y_pred_sqrt2)
```

```python
rfr_sqrt_predictions = rfr_sqrt2.predict(X).round()**2
```

```python
rfr_sqrt2_columns_names = ['rfr_sqrt_consent_fair_offer',
       'rfr_sqrt_consent_conditionally_fair_offer', 'rfr_sqrt_consent_unfair_offer',
       'rfr_sqrt_rejection_fair_offer', 'rfr_sqrt_rejection_conditionally_fair_offer',
       'rfr_sqrt_rejection_unfair_offer']

data_rfr_sqrt = pd.concat([data_rfr, pd.DataFrame(rfr_sqrt_predictions, columns=rfr_sqrt2_columns_names)], axis=1)
data_rfr_sqrt.head(5)
```

# Используем только некоторые предикторы для обучения модели

```python
# уберем признаки "Tolerance to uncertainty" и "Intolerance to uncertainty"
X = data[['Participant Private ID', 'Machiavellianism', 'Psychopathy',
       'Narcissism', 'Perspective-taking scale', 'Fantasy scale',
       'Empathic concern scale', 'Personal distress scale',
       'Interpersonal intolerance to uncertainty', 'genderRU-quantised',
       'DoBRU-year', 'educationRU-quantised']]

y = data[['consent_fair_offer',
       'consent_conditionally_fair_offer', 'consent_unfair_offer',
       'rejection_fair_offer', 'rejection_conditionally_fair_offer',
       'rejection_unfair_offer']]
```

```python
X[X['DoBRU-year'].isnull() == True]
```

```python
# далее повторим все предыдущие шаги
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

rfr_new = RandomForestRegressor(n_estimators=80, max_depth=2, max_leaf_nodes=2, random_state=0)
rfr_new.fit(X_train, y_train)

y_pred_new = rfr_new.predict(X_test).round()
mean_squared_error(y_test, y_pred_new)
```

```python
rfr_new2 = RandomForestRegressor(max_depth=2, max_leaf_nodes=2, random_state=0)
rfr_new2.fit(X_train, y_train)

y_pred_new2 = rfr_new2.predict(X_test).round()
mean_squared_error(y_test, y_pred_new2) # аналогичный результат при n_estimators=100
```

```python
rfr_new3 = RandomForestRegressor(n_estimators=137, max_depth=2, max_leaf_nodes=2, random_state=0)
rfr_new3.fit(X_train, y_train)

y_pred_new3 = rfr_new3.predict(X_test).round()
mean_squared_error(y_test, y_pred_new3) # аналогичный результат при n_estimators=136
```

Таким образом, убрав 2 признака, удалось **уменьшить ошибку модели** с 12.043859649122808 до 11.491228070175438, или **на 0.55263157894737**.

```python
rfr_new_predictions = rfr_new3.predict(X).round()

rfr_new_columns_names = ['rfr_consent_fair_offer',
       'rfr_consent_conditionally_fair_offer', 'rfr_consent_unfair_offer',
       'rfr_rejection_fair_offer', 'rfr_rejection_conditionally_fair_offer',
       'rfr_rejection_unfair_offer']

data_rfr_new = pd.concat([data, pd.DataFrame(rfr_new_predictions, columns=rfr_new_columns_names)], axis=1)
data_rfr_new.head(5)
```

```python
data_rfr_new.drop(data_rfr_new.tail(2).index, inplace=True)
```
