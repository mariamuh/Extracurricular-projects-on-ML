```python
# leave only necessary columns
triada_empathy = full_dataset[['Participant Private ID', 'Question Key', 'Response']]

# delete rows with system info
triada_empathy = triada_empathy[~triada_empathy['Question Key'].isin(['BEGIN QUESTIONNAIRE', 'END QUESTIONNAIRE', np.nan])]

# delete repeating rows
triada_empathy = triada_empathy.iloc[::2, :]
triada_empathy['Participant Private ID'] = triada_empathy['Participant Private ID'].astype(int)
```

```python
triada_empathy.shape[0]
```

```python
# shorten labels of tasks
triada_empathy.loc[:, 'Question Key'] = triada_empathy['Question Key'].apply(lambda x: x[8:-3])
```

```python
labels = {
    '1DD': 'M',
    '2DD': 'M',
    '3DD': 'M',
    '4DD': 'M',
    '5DD': 'P',
    '6DD': 'P',
    '7DD': 'P',
    '8DD': 'P',
    '9DD': 'N',
    '10DD': 'N',
    '11DD': 'N',
    '12DD': 'N',
    '1IRI': 'FS',
    '2IRI': 'EC',
    '3IRI': 'PT',
    '4IRI': 'EC',
    '5IRI': 'FS',
    '6IRI': 'PD',
    '7IRI': 'FS',
    '8IRI': 'PT',
    '9IRI': 'EC',
    '10IRI': 'PD',
    '11IRI': 'PT',
    '12IRI': 'FS',
    '13IRI': 'PD',
    '14IRI': 'EC',
    '15IRI': 'PT',
    '16IRI': 'FS',
    '17IRI': 'PD',
    '18IRI': 'EC',
    '19IRI': 'PD',
    '20IRI': 'EC',
    '21IRI': 'PT',
    '22IRI': 'EC',
    '23IRI': 'FS',
    '24IRI': 'PD',
    '25IRI': 'PT',
    '26IRI': 'FS',
    '27IRI': 'PD',
    '28IRI': 'PT'
}
```

```python
# add a column with labels of categories from 2 questionnaires (Dirty Dozen/Темная триада & Inerpersonal Reactivity Index/Эмпатия Дэвиса)
triada_empathy['Label of personal trial'] = triada_empathy['Question Key'].apply(lambda x: labels[x])
```

```python
triada_empathy.tail(28)
```

```python
# reverse answers 
triada_empathy['Response'] = triada_empathy.apply(lambda x: (6 - x.iloc[2]) if (x.iloc[1] in ['3IRI', '4IRI', '7IRI', '12IRI', '13IRI', '14IRI', '15IRI', '18IRI', '19IRI']) 
                                                  else x.iloc[2], axis=1)
```

```python
triada_empathy.tail(28)
```

```python
# grouped_triada_empathy = triada_empathy.groupby(['Participant Private ID', 'Label of personal trial']).sum()

# group by categoty
dataset = pd.pivot_table(triada_empathy, values=['Response'], index=['Participant Private ID'], columns=['Label of personal trial'], aggfunc=np.sum)

# indexes = dataset.index
# dataset = dataset.reset_index(drop=True).set_index(indexes.values)

dataset.columns = dataset.columns.levels[1]

# sort columns of the dataframe in correct order
dataset = dataset[['M', 'P', 'N', 'PT', 'FS', 'EC', 'PD']]
dataset.head(5)
```

```python
columns_names = {'M': 'Machiavellianism', 
                 'P': 'Psychopathy', 
                 'N': 'Narcissism', 
                 'PT': 'Perspective-taking scale', 
                 'FS': 'Fantasy scale', 
                 'EC': 'Empathic concern scale', 
                 'PD': 'Personal distress scale'}

dataset.columns = dataset.columns.map(lambda x: columns_names[x])
dataset.head(5)
```

```python
# drop title of columns
dataset = dataset.rename_axis(None, axis=1).reset_index()
dataset.head(5)
```
